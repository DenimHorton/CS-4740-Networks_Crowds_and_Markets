```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.zeros((4,4))
print(A)
```

```python
A[0,0:3] = [1/3,1/3,1/3]
A[1,3] = 1
A[2,3] = 1
A[3,0] = 1/2
A[3,3] = 1/2
print(A)
```

```python
timesteps = 16
xx = np.zeros((3,timesteps))
xx_new = np.zeros((4,timesteps))
xx[:,0] = [1,.5,0]
xx_new[:,0] = [1, 2/3, 1/3, 0]
```

```python
for i in range(1,timesteps) :
  xx[0,i] = (xx[0,i-1]+xx[2,i-1])/2
  xx[1,i] = xx[0,i-1]
  xx[2,i] = xx[1,i-1]
```

```python
for i in range(1,timesteps) :
  xx_new[:,i] = A@xx_new[:,i-1]
```

```python
plt.plot(xx_new.T)
```

```python
def opinion_example(timesteps,initial) :
  xx = np.zeros((3,timesteps))
  xx[:,0] = initial
  for i in range(1,timesteps) :
    xx[0,i] = (xx[0,i-1]+xx[2,i-1])/2
    xx[1,i] = xx[0,i-1]
    xx[2,i] = xx[1,i-1]
  plt.plot(xx.T)
  return xx[:,-1]
```

```python
opinion_example(50,[0,1,1])
```
