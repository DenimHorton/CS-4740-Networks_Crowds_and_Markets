```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
timesteps = 16
xx = np.zeros((3,timesteps))
xx[:,0] = [1,.5,0]
```

```python
for i in range(1,timesteps) :
  xx[0,i] = (xx[0,i-1]+xx[2,i-1])/2
  xx[1,i] = xx[0,i-1]
  xx[2,i] = xx[1,i-1]
```

```python
plt.plot(xx[0,:])
plt.plot(xx[1,:])
plt.plot(xx[2,:])
```

```python
plt.plot(xx.T)
```

```python
def opinion_example(timesteps,initial) :
  xx = np.zeros((3,timesteps))
  xx[:,0] = initial
  for i in range(1,timesteps) :
    xx[0,i] = (xx[0,i-1]+xx[2,i-1])/2
    xx[1,i] = xx[0,i-1]
    xx[2,i] = xx[1,i-1]
  plt.plot(xx.T)
  return xx[:,-1]
```

```python
opinion_example(50,[0,1,1])
```
